```python
b_ = list(b['Client Patient ID']) 
m_ = list(m['Client Patient ID'])
```

```python
cm = []
_b = [d.replace('-','') for d in b_]
for x in m_:
    if x in _b:
        cm.append(x)
```

```python
# Commom CPID between bridh and Marondera
len(cm)
```

```python
# Duplicated with drop last, keeping first
dups_1 = b[b.duplicated(['Client Patient ID'])]
dups_1.shape
```

```python
dups_1.head(2)
```

```python
# Duplicated keeping both dups for comparison purposes
dupds_2 = pd.concat(g for _, g in b.groupby("Client Patient ID") if len(g) > 1)
dupds_2.shape
```

```python
dupds_2.head(6)
```

```python
dupds_2[['Sample ID', 'Client Patient ID', 'Creator', 'Review State']]
```

```python
# Duplicated by the same creator
dups_3 = pd.concat(g for _, g in b.groupby(["Client Patient ID", "Creator"]) if len(g) > 1)
dups_3.shape
```

```python
dups_3.head(10).sort_values(by="Client Patient ID", ascending=False)[['Sample ID', 'Client Patient ID', 'Creator', 'Review State']]
```

```python
system_daemon  = dups_3.query('Creator == "system_daemon"')
system_daemon.shape
```

```python
system_daemon[['Sample ID', 'Client Patient ID', 'Creator', 'Review State']]
```

```python
system_ehr  = dups_3.query('Creator == "system_ehr"')
system_ehr.shape
```

```python
system_ehr[['Sample ID', 'Client Patient ID', 'Creator', 'Review State']]
```

```python
# we dont care about these: they have nothing to do with ehr
users = dups_3.query('Creator != "system_ehr" and Creator != "system_daemon"')
users.shape
```

```python
users[['Sample ID', 'Client Patient ID', 'Creator', 'Review State']]
```

```python
# Duplicated by the diff creators : all dups - dups by same creator
dups_4 = pd.concat([dupds_2, dups_3]).drop_duplicates(keep=False)
dups_4.shape
```

```python
dups_4 = dupds_2.append(dups_3).drop_duplicates(keep=False)
dups_4.shape
```
